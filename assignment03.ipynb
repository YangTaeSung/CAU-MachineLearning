```
import matplotlib.pyplot as plt
import numpy as np
```

# 1.Input points

```

path = "data.csv"
data = np.genfromtxt(path, delimiter=',')

x_data = data[:, 0]
y_data = data[:, 1]

plt.figure(figsize=(8, 8)) # 그래프 사이즈 조절
plt.scatter(x_data, y_data, c = 'black', alpha = 0.5) # alpha : 투명도
plt.title('Linear function')
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.show()
```

### [1] plot a set of points {(x(i), y(i))} that are loaded from 'data.csv' file (in black color)



```
thetaZero = 0.0
thetaOne = 0.0

cycle = 3000
learningLate = 0.01
```

* 세타값 초기화(learning 시작점)
* Optimization cycle과 learning late 설정

```
iForGraph = []
jFor3D = []
newThetaOne = []
newThetaZero = []
noiseNum = len(x_data)
```

* Plotting을 위한 각종 데이터 저장 변수 정의

```
for i in range(cycle):

  h = thetaZero + thetaOne * x_data
  j = np.sum((h - y_data) ** 2) / (2 * noiseNum)

  thetaZero = thetaZero - learningLate / noiseNum * np.sum((thetaOne * x_data - y_data + thetaZero))
  thetaOne = thetaOne - learningLate / noiseNum * np.sum((thetaOne * x_data - y_data + thetaZero) * x_data)
  
  iForGraph.append(i)
  jFor3D.append(j)
  newThetaOne.append(thetaOne)
  newThetaZero.append(thetaZero)
  
  if i % 100 == 0:
        print('cycle : {:10d} cost: {:10f} thetaZero: {:10f} thetaOne: {:10f}'.format(i, j,thetaZero, thetaOne))

finalY = thetaOne * x_data + thetaZero 
```

* Gradient descent algorithm 
* list.append 함수를 사용하여 optimization 단계마다 각각의 값 저장
* 100cycle 주기로 출력
* Optimal한 linear함수 finalY 정의

# 2.Linear regression result

```
plt.figure(figsize=(8, 8)) # 그래프 사이즈 조절
plt.scatter(x_data, y_data, c = 'black', alpha = 0.5) # alpha : 투명도
plt.plot(x_data, finalY, "RED")
plt.title('Linear function')
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.show()
```
